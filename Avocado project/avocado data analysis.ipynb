```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
avo = avo.drop(['ID','Date'], axis =1)
```

```python
avo.head(3)
```

```python
avo['type']=avo['type'].map({'conventional':1,'organic':0})
```

```python
avo.head(3)
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x=avo['year'],y=avo['Average Price'])
plt.savefig('Yearly Avocado Prices')
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
sns.scatterplot(x=avo['Average Price'],y=avo['Total Volume'])
plt.title('Average Price vs Volume')
plt.xlabel('Avocado Average Price')
plt.ylabel('Avocado Total Volume')

plt.subplot(1,2,2)
sns.scatterplot(x=avo['Average Price'],y=avo['Total Bags'])
plt.title('Average Price vs Bags')
plt.xlabel('Avocado Average Price')
plt.ylabel('Avocado Total Bags')
```

```python
avo.corr()
```

```python
sns.heatmap(avo.corr(), annot=True, cmap = 'BuGn')
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
sns.barplot(x=avo['year'],y=avo['Total Volume'])

plt.figure(figsize=(15,5))
plt.subplot(1,2,2)
sns.barplot(x=avo['year'],y=avo['Total Bags'])
```
